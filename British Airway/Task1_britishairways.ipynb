```python
base_url = "https://www.airlinequality.com/airline-reviews/british-airways"
pages = 10
page_size = 100

reviews = []

# for i in range(1, pages + 1):
for i in range(1, pages + 1):

    print(f"Scraping page {i}")

    # Create URL to collect links from paginated data
    url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"

    # Collect HTML data from this page
    response = requests.get(url)

    # Parse content
    content = response.content
    parsed_content = BeautifulSoup(content, 'html.parser')
    for para in parsed_content.find_all("div", {"class": "text_content"}):
        reviews.append(para.get_text())

    print(f"   ---> {len(reviews)} total reviews")
```

```python
df = pd.DataFrame()
df["reviews"] = reviews
df.head()
```

## Step 2: Clean and Analyze the Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import string
from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from textblob import TextBlob
import seaborn as sns
import spacy
from collections import Counter
```

```python
# Load the data
df = pd.read_csv("data/BA_reviews.csv", index_col=0)
print(f"Dataset shape: {df.shape}")
print(df.head())
```

```python
# Data Cleaning
def clean_text(text):
    """Clean and prepare text for analysis"""
    # Convert to lowercase
    text = str(text).lower()

    # Remove verification tags
    text = re.sub(r'✅ trip verified \|', '', text)
    text = re.sub(r'not verified \|', '', text)

    # Remove punctuation
    text = re.sub(r'[^\w\s]', ' ', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
# Apply cleaning to the reviews
df['cleaned_review'] = df['reviews'].apply(clean_text)
print("\nCleaned Reviews:")
print(df['cleaned_review'].head())
```

```python
# Load spaCy model
print("Loading spaCy model...")
nlp = spacy.load("en_core_web_sm")
```

```python
# Create custom stopwords specific to airline reviews
custom_stopwords = ['british', 'airways', 'airline', 'flight', 'ba', 'plane', 'passenger', 'seat', 'airport']
```

```python
# Function to preprocess text using spaCy
def preprocess_text(text):
    """Preprocess text using spaCy for tokenization, stopword removal, and lemmatization"""
    # Process text with spaCy
    doc = nlp(text)

    # Extract tokens
    # - Keep only content words (no stop words)
    # - Remove punctuation and spaces
    # - Lemmatize tokens
    # - Remove short words (length < 3)
    tokens = [token.lemma_.lower() for token in doc
              if not token.is_stop
              and not token.is_punct
              and not token.is_space
              and len(token.text) > 2
              and token.lemma_.lower() not in custom_stopwords]

    return tokens
```

```python
# Apply preprocessing and create tokenized text
df['tokens'] = df['cleaned_review'].apply(preprocess_text)
```

```python
# Create a column with the tokens joined back into a string for certain analyses
df['processed_text'] = df['tokens'].apply(lambda x: ' '.join(x))
```

```python
print("\nProcessed text examples:")
print(df['processed_text'].head())
```

```python
# Extract and analyze verified vs non-verified reviews
df['verified'] = df['reviews'].apply(lambda x: 1 if '✅ Trip Verified' in x else 0)
verified_count = df['verified'].sum()
print(f"\nVerified reviews: {verified_count} ({verified_count/len(df)*100:.1f}%)")
print(f"Non-verified reviews: {len(df)-verified_count} ({(len(df)-verified_count)/len(df)*100:.1f}%)")
```

```python
# Sentiment Analysis
def get_sentiment(text):
    analysis = TextBlob(text)
    # Return polarity score (-1 to 1)
    return analysis.sentiment.polarity
```

```python
# Apply sentiment analysis
print("Performing sentiment analysis...")
df['sentiment'] = df['cleaned_review'].apply(get_sentiment)
```

```python
# Create sentiment categories
df['sentiment_category'] = pd.cut(
    df['sentiment'],
    bins=[-1.1, -0.1, 0.1, 1.1],
    labels=['Negative', 'Neutral', 'Positive']
)
```

```python
# Calculate sentiment distribution
sentiment_counts = df['sentiment_category'].value_counts()
print("\nSentiment Distribution:")
print(sentiment_counts)
print(f"Average sentiment score: {df['sentiment'].mean():.3f}")
```

```python
# Compare sentiment between verified and non-verified reviews
verified_sentiment = df[df['verified'] == 1]['sentiment'].mean()
nonverified_sentiment = df[df['verified'] == 0]['sentiment'].mean()
print(f"\nAverage sentiment (verified reviews): {verified_sentiment:.3f}")
print(f"Average sentiment (non-verified reviews): {nonverified_sentiment:.3f}")
```

```python
# Topic Modeling
# Define the number of topics
n_topics = 5

# Create a document-term matrix
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
doc_term_matrix = vectorizer.fit_transform(df['processed_text'])

# Create and fit the LDA model
print("Fitting LDA model...")
lda = LatentDirichletAllocation(
    n_components=n_topics,
    random_state=42,
    max_iter=20,  # Increased iterations for better convergence
    learning_method='online',  # Faster for larger datasets
    n_jobs=-1  # Use all available cores
)
lda.fit(doc_term_matrix)
```

```python
# Function to print top words for each topic
def print_top_words(model, feature_names, n_top_words):
    topics = []
    for topic_idx, topic in enumerate(model.components_):
        top_words = [feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]
        topics.append((topic_idx, top_words))
        print(f"Topic #{topic_idx}: {', '.join(top_words)}")
    return topics
```

```python
# Get features names from vectorizer
feature_names = vectorizer.get_feature_names_out()

print("\nTopics identified in British Airways reviews:")
topics = print_top_words(lda, feature_names, 10)
```

```python
# Create a DataFrame for topic analysis
topic_matrix = lda.transform(doc_term_matrix)
df['dominant_topic'] = topic_matrix.argmax(axis=1)
```

```python
# Analyze sentiment by topic
topic_sentiment = df.groupby('dominant_topic')['sentiment'].mean().reset_index()
print("\nAverage sentiment by topic:")
print(topic_sentiment)
```

```python
# Save the topics with their keywords
topic_keywords = pd.DataFrame(columns=['Topic', 'Keywords'])
for topic_id, words in topics:
    topic_keywords = pd.concat([topic_keywords,
                               pd.DataFrame({'Topic': [f"Topic {topic_id}"],
                                           'Keywords': [', '.join(words)]})],
                              ignore_index=True)
topic_keywords.to_csv('topic_keywords.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
sentiment_counts = df['sentiment_category'].value_counts()
labels = sentiment_counts.index.tolist()
sizes = sentiment_counts.values.tolist()

# Setup
fig, ax = plt.subplots(figsize=(10, 8))
colors = ['#d1c4d2', '#c6f6d5', '#fed7d7']  # Adjust as needed
explode = [0.05] * len(sizes)  # Explode all slightly for emphasis

# Define correct autopct function using index from enumerate
def make_autopct(values, labels):
    def my_autopct(pct, values=values, labels=labels):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        label = labels[np.where(np.array(values) == val)[0][0]]  # Get correct label based on value
        return f"{pct:.1f}%\n({label})"  # Percentage and label
    return my_autopct

# Plot
wedges, texts, autotexts = ax.pie(
    sizes,
    explode=explode,
    colors=colors,
    startangle=90,
    autopct=lambda pct: make_autopct(sizes, labels)(pct),  # Use the corrected autopct function
    pctdistance=0.65,
    shadow=True
)

# Style text inside pie
for autotext in autotexts:
    autotext.set_fontsize(11)
    autotext.set_color('black')
    autotext.set_fontweight('bold')

ax.set_title("British Airways Review Sentiment Breakdown", fontsize=16, fontweight='bold')
ax.set_aspect('equal')
plt.tight_layout()
plt.savefig('sentiment_pie_corrected.png', dpi=300, bbox_inches='tight', transparent=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import FancyBboxPatch

# Sentiment data
sentiment_counts = df['sentiment_category'].value_counts()
labels = ['Positive', 'Neutral', 'Negative']
sizes = [sentiment_counts.get(label, 0) for label in labels]
colors = ['#c6f6d5', '#d1c4d2', '#fed7d7']
total = sum(sizes)
percentages = [s / total * 100 for s in sizes]

# Metrics
avg_sentiment = df['sentiment'].mean()
total_reviews = len(df)
pos_percentage = (df['sentiment_category'] == 'Positive').mean() * 100
neg_percentage = (df['sentiment_category'] == 'Negative').mean() * 100
verified_percentage = df['verified'].mean() * 100

# Create larger layout
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 8), gridspec_kw={'width_ratios': [4, 2]})
fig.suptitle('British Airways Review Summary', fontsize=20, fontweight='bold')

# --- Donut Chart (larger) ---
wedges, _ = ax1.pie(
    sizes, startangle=90, colors=colors, labels=None,
    wedgeprops={'width': 0.50, 'edgecolor': 'white'}
)
for i, wedge in enumerate(wedges):
    angle = (wedge.theta2 + wedge.theta1) / 2
    x, y = np.cos(np.deg2rad(angle)), np.sin(np.deg2rad(angle))
    ax1.text(x * 0.7, y * 0.7, f"{labels[i]}\n{percentages[i]:.1f}%",
             ha='center', va='center', fontsize=13, fontweight='bold')

# Centre label
ax1.text(0, 0.1, 'Avg Sentiment', ha='center', va='center', fontsize=12)
ax1.text(0, -0.1, f"{avg_sentiment:.2f}", ha='center', va='center',
         fontsize=17, fontweight='bold', color='navy')
ax1.set_title('Sentiment Distribution', fontsize=16, fontweight='bold')

# --- Centre-aligned Key Metrics with colour cards ---
ax2.axis('off')
ax2.set_title('Key Metrics', fontsize=16, fontweight='bold', loc='center', pad=20)

# Metric data (label, value, box colour, text colour)
metrics = [
    ("Total Reviews",     f"{total_reviews:,}",        '#d0e6ff', '#003366'),
    ("Avg Sentiment",     f"{avg_sentiment:.3f}",      '#cfefff', '#1f77b4'),
    ("Positive Reviews",  f"{pos_percentage:.1f}%",    '#c6f6d5', '#276749'),
    ("Negative Reviews",  f"{neg_percentage:.1f}%",    '#fed7d7', '#c53030'),
    ("Verified Reviews",  f"{verified_percentage:.1f}%",'#e9d8fd', '#6b46c1')
]

# Draw coloured metric cards
for i, (label, value, bgcolor, textcolor) in enumerate(metrics):
    y = 0.8 - i * 0.18
    height = 0.12
    width = 0.9
    # Draw background box
    box = FancyBboxPatch((0.05, y), width, height,
                         boxstyle="round,pad=0.02", ec="none", fc=bgcolor, transform=ax2.transAxes)
    ax2.add_patch(box)
    # Add text on top
    ax2.text(0.1, y + 0.06, label, fontsize=12, fontweight='bold',
             color='black', transform=ax2.transAxes, ha='left', va='center')
    ax2.text(0.93, y + 0.06, value, fontsize=13, fontweight='bold',
             color=textcolor, transform=ax2.transAxes, ha='right', va='center')

# Final layout
plt.tight_layout(pad=5)
plt.savefig("sentiment_summary_beautified.png", dpi=300, bbox_inches='tight')
plt.show()
```
